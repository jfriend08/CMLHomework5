## Compare loss function
- Plot hinge loss and the huberized hinge loss function
- So as following the code below, we can see huberized hinge loss will be smoothed at [1-h, 1+h] range


````python
%matplotlib inline

import gradientDescent as gd
import numpy as np
import random, sys
import matplotlib.pyplot as plt
from sklearn import preprocessing

yt = np.arange(0., 2, 0.005)
out1 = map(gd.lossHinge, yt)
out2 = map(gd.lossHuberHinge, yt)
plt.plot(yt, out1, 'r--', yt, out2, 'b--')
plt.title('Loss function')
plt.ylabel('Value')
plt.xlabel('yt')
plt.show()
````

## Proof differentiable and Lipschitz-continuous
-- Please see the left part of the figure. To prove huberized hinge loss function is differentiable among all points yt I take the deliverative of loss function for yt. The slope of left point at very close to 1-h/1+h is the same as right point close to 1-h/1+h, and therefore this huberized hinge loss function is a differentiable function. 

-- Please see the right part of the figure. This function is 0-Lips and 1-Lips when yt > 1+h and yt < 1-h respectively. And although when yt at the range [1-h, 1+h] may not looks like Lips when first glance, but since yt is within a certain range, and it's still bounded in a certain range. 

-- Since Lipschitz is varied at different range, this function is not Lipschitz-continuous

- Make Grad_checker a generic, and to check whether compute_obj and compute_grad are correct


````python
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(-1, 1), copy=True)
x = np.array([np.zeros(5), np.zeros(5), np.ones(5), np.ones(5), np.zeros(5), np.zeros(5)])
x = min_max_scaler.fit_transform(x)
y = np.array([-1, -1, 1, 1, -1, -1])

mygd = gd.gradientDescent(x, y)
w = np.array([1, 1, 0, 1, 0])
mygd.compute_obj(w)
mygd.compute_grad(w)

for i in xrange(5):
  w = np.random.randint(100, size=(1, 5)).flatten()
  print i, "th checking: error sum\n", mygd.grad_checker(mygd.compute_obj, mygd.compute_grad, w)
  print "----------------------"
````

## Data points
- Generate the data points with proper numbers (n) and dimension(d) the way I want
- Normalize the data
- Split datas into training and testing set


````python
import points as pt
from sklearn.cross_validation import train_test_split
X, y = pt.dataset_fixed_cov(700, 10, 3) #n, dim, overlapped dist
print "X.shape", X.shape
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(-1, 1), copy=True)
X = min_max_scaler.fit_transform(X)
rng = np.random.RandomState(19850920)
permutation = rng.permutation(len(X))
X, y = X[permutation], y[permutation]
train_X, test_X, train_y, test_y = train_test_split(X, y, train_size=0.1, random_state=2010)
pt.plotPCA(X, y)
````

## Run gradient descent
- Plot the objective function vs the iterations
- As ita, which is the step size, increases, there is some point(0.1 \* 1.1\**20) the accuracy vibrate between 98% to 2%

````python
import svm
n, dim = X.shape
mysvm = svm.svm()
w = np.zeros(dim)

ita = 0.1 * 1.1**10

wIter, w, iterCount = mysvm.fit(train_X, train_y, w, method="gd", compute_obj="Default", compute_grad="Default", maxiter=50, ita=ita, c=1)
# def getAccuracy(w, X, y):
#     predict = np.apply_along_axis(lambda eachx: np.dot(w,eachx), 1, X)
#     diff = predict*y
#     return len(np.where(diff>0)[0])/float(len(X))
# accuracy = map(lambda w: getAccuracy(w, train_X, train_y), wIter)
accuracy = gd.getAccuracyOverIteration(wIter, X, y)
mygd = gd.gradientDescent(train_X, train_y)
objF = map(mygd.compute_obj,wIter)

plt.figure(1)
plt.subplot(211)
plt.plot(accuracy, 'r--')
plt.ylabel('Prediction Accuracy')
plt.subplot(212)
plt.plot(objF, 'r--')
plt.ylabel('objF value')
plt.xlabel('Each Iteration')
plt.show()
print "last value from objective function:", objF[-1]
````

## Backtracking line-searching
- Below is the equation for searching the optimal step

```
t = 1
a = 1
b = 0.9
dw = np.ones(len(w))
while compute_obj(w+t*dw) > compute_obj(w) + a*t*np.dot(compute_grad(w),dw):
    t *= b

```
- As we can see here, although line search got a small step update. This new step gave us a stable objective function that doesn't jump dramatically, while yielding comparable prediction accuracy

````python
wIter, w, iterCount = mysvm.fit(train_X, train_y, w, method="gd", compute_obj="Default", compute_grad="Default", maxiter=50, ita=ita, c=1, Step_backtrack=True)
accuracy = gd.getAccuracyOverIteration(wIter, X, y)
mygd = gd.gradientDescent(train_X, train_y)
objF = map(mygd.compute_obj,wIter)

plt.figure(1)
plt.subplot(211)
plt.plot(accuracy, 'r--')
plt.ylabel('Prediction Accuracy')
plt.subplot(212)
plt.plot(objF, 'r--')
plt.ylabel('objF value')
plt.xlabel('Each Iteration')
plt.show()
print "last value from objective function:", objF[-1]
````

## Grid-search on C
- Here I done the grid-search on ita and c, and I collected the min-performance(defined as mis-classficatin rate) over each test cases aiming to compare the best of their worst performance. 
- Looks like ita and ctest play a crucial and minor role respectively here. As lone as ita is too big, there is always a problem to have a good performance. But if itea now is at a reasonable size but C is too big, then we can see there is a bad performance as well

````python
import svm
n, dim = X.shape
w = np.zeros(dim)
mysvm = svm.svm()
searchResult = []
for itatest in [0.5, 0.3, 0.1, 0.05, 0.01]:
    for ctest in [1000, 100, 10, 5, 1, 0.8, 0.6, 0.4, 0.2, 0.01]:
        wIter, w, iterCount = mysvm.fit(train_X, train_y, w, method="gd", compute_obj="Default", compute_grad="Default", maxiter=50, ita=itatest, c=ctest, stopMethod="performance")
        accuracy = np.amin(gd.getAccuracyOverIteration(wIter, X, y))
#         print "--------", "accuracy:", accuracy, "--------"
        searchResult.append([itatest, ctest, accuracy, w] )
        
# print searchResult
# We extract just the scores
scores = [x[2] for x in searchResult]
print np.array(scores).shape
scores = np.array(scores).reshape(10, 5)
plt.matshow(scores)
plt.colorbar()
````

## Large point
- 10^7 points are hard to handle at local and slow things down even for generate point. So here I generate 10^5 points on 10-dimention space


````python
import points as pt
from sklearn import preprocessing
from sklearn.cross_validation import train_test_split
X, y = pt.dataset_fixed_cov(100000, 10, 3) #n, dim, overlapped dist
Xsmall, ysmall = pt.dataset_fixed_cov(10000, 10, 3) #n, dim, overlapped dist
print "X.shape", X.shape
print "Xsmall.shape", Xsmall.shape
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(-1, 1), copy=True)
X = min_max_scaler.fit_transform(X)
rng = np.random.RandomState(19850920)
permutation = rng.permutation(len(X))
X, y = X[permutation], y[permutation]
train_X, test_X, train_y, test_y = train_test_split(X, y, train_size=0.1, random_state=2010)
train_Xsmall, test_Xsmall, train_ysmall, test_ysmall = train_test_split(Xsmall, ysmall, train_size=0.1, random_state=2010)
pt.plotPCA(X, y)
````

## Compare speed
- We have data set for 10^5 and 10^4 points, and the both run batch gradient descent at the same condition.
- The 10^5 data set need 24 second to finish 50 iteration, while 10^4 data set tooks only 2.44 second. 
- So the slow-down became very noticiable then the data set got big, and I think it is because batch gradient descent need to consider all the slope from all points in data to update its new w. This might eliminate noises within data and take the most objective step, but this simply not working if the data size is huge.

````python
import svm
import time

n, dim = X.shape
w = np.zeros(dim)
mysvm = svm.svm()
start_time = time.time()
wIter, w, iterCount = mysvm.fit(train_X, train_y, w, method="gd", compute_obj="Default", compute_grad="Default", maxiter=50, ita=0.1, c=1)
print "iterCount", iterCount
print("--- %s seconds ---" % (time.time() - start_time))

n, dim = Xsmall.shape
w = np.zeros(dim)
mysvm = svm.svm()
start_time = time.time()
wIter, w, iterCount = mysvm.fit(train_Xsmall, train_ysmall, w, method="gd", compute_obj="Default", compute_grad="Default", maxiter=50, ita=0.1, c=1)
print "iterCount", iterCount
print("--- %s seconds ---" % (time.time() - start_time))
````
