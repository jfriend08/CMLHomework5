## Compare loss function

````python
%matplotlib inline

import gradientDescent as gd
import numpy as np
import random, sys
import matplotlib.pyplot as plt
from sklearn import preprocessing

yt = np.arange(0., 2, 0.005)
out1 = map(gd.lossHinge, yt)
out2 = map(gd.lossHuberHinge, yt)
plt.plot(yt, out1, 'r--', yt, out2, 'b--')
plt.title('Loss function')
plt.ylabel('Value')
plt.xlabel('yt')
plt.show()
````

## Proof differentiable and Lipschitz-continuous
-- Please see the left part of the figure. To prove huberized hinge loss function is differentiable among all points yt, I take the deliverative of loss function for yt. The slope of left point at very close to 1-h/1+h is the same as right point close to 1-h/1+h, and therefore this huberized hinge loss function is a differentiable function. 

-- Please see the right part of the figure. This function is 0-Lips and 1-Lips when yt > 1+h and yt < 1-h respectively. And although when yt at the range [1-h, 1+h] may not looks like Lips when first glance, but since yt is within a certain range, and it's still bounded in a certain range. 

-- Since Lipschitz is always bounded for any range, and therefore this function is not Lipschitz-continuous

## Backtracking line-searching
- Below is the equation for searching the optimal step

```
t = 1
a = 1
b = 0.9
dw = np.ones(len(w))
while compute_obj(w+t*dw) > compute_obj(w) + a*t*np.dot(compute_grad(w),dw):
    t *= b

```
- As we can see here, although line search got a small step update. This new step gave us a stable objective function that doesn't jump dramatically, while yielding comparable prediction accuracy

````python
wIter, w, iterCount = mysvm.fit(train_X, train_y, w, method="gd", compute_obj="Default", compute_grad="Default", maxiter=50, ita=ita, c=1, Step_backtrack=True)
accuracy = gd.getAccuracyOverIteration(wIter, X, y)
mygd = gd.gradientDescent(train_X, train_y)
objF = map(mygd.compute_obj,wIter)

plt.figure(1)
plt.subplot(211)
plt.plot(accuracy, 'r--')
plt.ylabel('Prediction Accuracy')
plt.subplot(212)
plt.plot(objF, 'r--')
plt.ylabel('objF value')
plt.xlabel('Each Iteration')
plt.show()
print "last value from objective function:", objF[-1]
````

## Grid-search on C and ita
- Here I done the grid-search on ita and c, and I collected the min-performance(defined as mis-classficatin rate) over each test cases aiming to compare the best of their worst performance. 
- Looks like ita and ctest play a crucial and minor role respectively here. As lone as ita is too big, there is always a problem to have a good performance. But if itea now is at a reasonable size but C is too big, then we can see there is a bad performance as well

````python
import svm
n, dim = X.shape
w = np.zeros(dim)
mysvm = svm.svm()
searchResult = []
for itatest in [0.5, 0.3, 0.1, 0.05, 0.01]:
    for ctest in [1000, 100, 10, 5, 1, 0.8, 0.6, 0.4, 0.2, 0.01]:
        wIter, w, iterCount = mysvm.fit(train_X, train_y, w, method="gd", compute_obj="Default", compute_grad="Default", maxiter=50, ita=itatest, c=ctest, stopMethod="performance")
        accuracy = np.amin(gd.getAccuracyOverIteration(wIter, X, y))
#         print "--------", "accuracy:", accuracy, "--------"
        searchResult.append([itatest, ctest, accuracy, w] )
        
# print searchResult
# We extract just the scores
scores = [x[2] for x in searchResult]
print np.array(scores).shape
scores = np.array(scores).reshape(10, 5)
plt.matshow(scores)
plt.colorbar()
````

## Implementation of Stochastic Gradient Descent
- Instead of considering all points together to make one update for w during each iteration. SGD just perform each update for each invidual data point, taking the fastest dirction toward the encounter point. Furthermore, to previent data-sequenc-related issue, data shuffl is performed during each iteration(epoch)
- Before really compare the efficiency when dealing with large data set, below are experiments to understand its basic properities
- What w will return is the averaged w for all w during this iteration. And I would like to see the differences between averagedW and allW. 
- For averagedW, the prediction accuracy got very similar value at the first iteration, which is the same as gradient descent. One interesting result is that the objective function can smoothly being improve over each iteration.
- Taking the more detail resolution looking at each w over all iteration, we can see there are really lots of jumping for prediction accuracy and objective function. I think this is mainly due to sgd is taking the fastest dirction toward each single point, but might not be the best route, which yielding bad performance or objective value. As you can see each jumping point corresponding to the same point at accuracy and objective functin figure.
- I still implemented the same three stopping methods as gradient descent, and I realized why "optimize" method is not necessary because basically each of the update will yield very different objective valuse, or at least bigger than 0.001 difference, and therefore "optimize" method is almost just the same as "maxIter" method

````python
import svm
import points as pt
import numpy as np
from sklearn import preprocessing
from sklearn.cross_validation import train_test_split

Xsmall, ysmall = pt.dataset_fixed_cov(100, 10, 3) #n, dim, overlapped dist
train_Xsmall, test_Xsmall, train_ysmall, test_ysmall = train_test_split(Xsmall, ysmall, train_size=0.5, random_state=2010)

mysvm = svm.svm()
n, dim = train_Xsmall.shape
w = np.zeros(dim)
mygd = gd.gradientDescent(train_Xsmall, train_ysmall)
w = np.zeros(dim)
averagedwIter, allw, w, iterCount = mysvm.fit(train_Xsmall, train_ysmall, w, method="sgd", compute_obj="Default", compute_grad="Default", maxiter=3, ita=0.11, stopMethod="optimize")

accuracyAveragedW = gd.getAccuracyOverIteration(averagedwIter, train_Xsmall, train_ysmall)
accuracyAllW = gd.getAccuracyOverIteration(allw, train_Xsmall, train_ysmall)
objectiveAveragedW = map(mygd.compute_obj, averagedwIter)
objectiveAllW = map(mygd.compute_obj, allw)

plt.figure(1)
plt.subplot(221)
plt.plot(accuracyAveragedW, 'r--')
plt.ylabel('Prediction Accuracy')
plt.subplot(222)
plt.plot(accuracyAllW, 'r--')


plt.subplot(223)
plt.plot(objectiveAveragedW, 'r--')
plt.ylabel('Objective value')
plt.xlabel('Averaged W')

plt.subplot(224)
plt.plot(objectiveAllW, 'r--')
plt.xlabel('allW')
plt.show()
````

## Ita0 and t0
- This is the part to play with step size ita during each of iteration, and the definition of ita is defined below:
```
ita = ita0/(iterationCount+t0)
```
- Below is the section to have fixed t0 but varied step size ita. What we can noticed from this experiment is that large step size is problematic, like the case of ita=1, that we simply not able to approach minima with this large step size

````python
mysvm = svm.svm()
n, dim = train_Xsmall.shape
w = np.zeros(dim)
mygd = gd.gradientDescent(train_Xsmall, train_ysmall)
w = np.zeros(dim)
averagedwIter1, allw1, w1, iterCount1 = mysvm.fit(train_Xsmall, train_ysmall, w, method="sgd", compute_obj="Default", stopMethod="optimize"
                                              , compute_grad="Default", maxiter=50, ita=1, itaOverIteration=True, tnot=1)
averagedwIter2, allw2, w2, iterCount2 = mysvm.fit(train_Xsmall, train_ysmall, w, method="sgd", compute_obj="Default", stopMethod="optimize"
                                              , compute_grad="Default", maxiter=50, ita=0.5
                                              , itaOverIteration=True, tnot=1)
averagedwIter3, allw3, w3, iterCount3 = mysvm.fit(train_Xsmall, train_ysmall, w, method="sgd", compute_obj="Default", stopMethod="optimize"
                                              , compute_grad="Default", maxiter=50, ita=0.25, itaOverIteration=True, tnot=1)
averagedwIter4, allw4, w4, iterCount4 = mysvm.fit(train_Xsmall, train_ysmall, w, method="sgd", compute_obj="Default", stopMethod="optimize"
                                              , compute_grad="Default", maxiter=50, ita=0.1, itaOverIteration=True, tnot=1)

objectiveAllW1 = map(mygd.compute_obj, allw1)
objectiveAllW2 = map(mygd.compute_obj, allw2)
objectiveAllW3 = map(mygd.compute_obj, allw3)
objectiveAllW4 = map(mygd.compute_obj, allw4)

plt.figure(1)
plt.subplot(221)
plt.plot(objectiveAllW1, 'r--')
plt.subplot(222)
plt.plot(objectiveAllW2, 'r--')
plt.subplot(223)
plt.plot(objectiveAllW3, 'r--')
plt.subplot(224)
plt.plot(objectiveAllW4, 'r--')
plt.show()
````


- With the knowledge of previous experiment that seems ita=0.1 play a better role on minimizing objective function, below is the section to have fixed ita=0.1 and various t0. ## Ita0 and t0
- This is the part of warm-up function that selects the best ita0 and t0

````python
import sys
mysvm = svm.svm()
n, dim = train_Xsmall.shape
w = np.zeros(dim)
mygd = gd.gradientDescent(train_Xsmall, train_ysmall)

maxVal = sys.float_info.min
bestIta0 = None
bestT0 = None
for ita0 in [0.5, 0.2, 0.1, 0.05, 0.02, 0.01]:
    for t0 in [10, 5, 2, 1, 0.5, 0.2, 0.1]:
        averagedwIter1, allw1, w1, iterCount1 = mysvm.fit(train_Xsmall, train_ysmall, w, method="sgd", compute_obj="Default", stopMethod="performance"
                                              , compute_grad="Default", maxiter=50, ita=ita0, itaOverIteration=True, tnot=t0)
        accuracyAllW = np.amax(gd.getAccuracyOverIteration(allw1, train_Xsmall, train_ysmall))
        print "accuracyAllW", accuracyAllW, "maxVal", maxVal
        if maxVal < accuracyAllW:
            print "----------------- updating -----------------"
            maxVal = accuracyAllW
            bestIta0 = ita0
            bestT0 = t0
print "Search Done", "maxVal:", maxVal, "bestIta0", bestIta0, "bestT0", bestT0
        
````
